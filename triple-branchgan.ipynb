```python
import os
import glob
import random
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torchvision.utils as vutils
from torchvision.models import vgg16
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
import lpips
import piq
from tqdm import tqdm
import matplotlib.pyplot as plt

# Dataset loader
class LowLightDataset(Dataset):
    def __init__(self, low_dir, high_dir, transform=None):
        self.low_images = sorted(glob.glob(os.path.join(low_dir, '*')))
        self.high_images = sorted(glob.glob(os.path.join(high_dir, '*')))
        self.transform = transform

    def __len__(self):
        return len(self.low_images)

    def __getitem__(self, idx):
        low_img = Image.open(self.low_images[idx]).convert('RGB')
        high_img = Image.open(self.high_images[idx]).convert('RGB')
        if self.transform:
            low_img = self.transform(low_img)
            high_img = self.transform(high_img)
        return low_img, high_img

# Triple-Branch Generator
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()

        # Branch 1: Structure
        self.structure_branch = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1), nn.ReLU()
        )

        # Branch 2: Illumination
        self.illum_branch = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1), nn.ReLU()
        )

        # Branch 3: Detail (e.g., noise handling)
        self.detail_branch = nn.Sequential(
            nn.Conv2d(3, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, 1, 1), nn.ReLU()
        )

        # Channel attention (on fused features)
        self.channel_attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(192, 48, 1), nn.ReLU(),
            nn.Conv2d(48, 192, 1), nn.Sigmoid()
        )

        # Fusion + Output layer
        self.output_layer = nn.Sequential(
            nn.Conv2d(192, 64, 3, 1, 1), nn.ReLU(),
            nn.Conv2d(64, 3, 3, 1, 1), nn.Tanh()
        )

    def forward(self, x):
        s_feat = self.structure_branch(x)
        i_feat = self.illum_branch(x)
        d_feat = self.detail_branch(x)
        fused = torch.cat([s_feat, i_feat, d_feat], dim=1)
        att = self.channel_attention(fused)
        fused = fused * att
        return self.output_layer(fused)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 1), nn.LeakyReLU(0.2),
            nn.Conv2d(64, 128, 4, 2, 1), nn.InstanceNorm2d(128), nn.LeakyReLU(0.2),
            nn.Conv2d(128, 256, 4, 2, 1), nn.InstanceNorm2d(256), nn.LeakyReLU(0.2),
            nn.Conv2d(256, 512, 4, 2, 1), nn.InstanceNorm2d(512), nn.LeakyReLU(0.2),
            nn.Conv2d(512, 1, 4, 1, 0), nn.Sigmoid()
        )

    def forward(self, x):
        return self.main(x)

class VGGPerceptual(nn.Module):
    def __init__(self):
        super(VGGPerceptual, self).__init__()
        vgg = vgg16(pretrained=True).features[:16]
        for param in vgg.parameters():
            param.requires_grad = False
        self.vgg = vgg.cuda()

    def forward(self, x):
        return self.vgg(x)

def high_pass_filter(img):
    return img - nn.functional.avg_pool2d(img, kernel_size=3, stride=1, padding=1)

def edge_loss(gen_img, gt_img):
    sobel_x = torch.tensor([[-1,0,1],[-2,0,2],[-1,0,1]], dtype=torch.float32, device=gen_img.device).view(1,1,3,3)
    sobel_y = sobel_x.transpose(2,3)
    gx1 = nn.functional.conv2d(gen_img, sobel_x.expand(3, -1, -1, -1), groups=3, padding=1)
    gy1 = nn.functional.conv2d(gen_img, sobel_y.expand(3, -1, -1, -1), groups=3, padding=1)
    gx2 = nn.functional.conv2d(gt_img, sobel_x.expand(3, -1, -1, -1), groups=3, padding=1)
    gy2 = nn.functional.conv2d(gt_img, sobel_y.expand(3, -1, -1, -1), groups=3, padding=1)
    return nn.functional.l1_loss(gx1, gx2) + nn.functional.l1_loss(gy1, gy2)

def compute_metrics(model, dataloader):
    model.eval()
    total_psnr = total_ssim = total_fsim = total_mae = 0
    count = 0
    with torch.no_grad():
        for low_img, high_img in dataloader:
            low_img = low_img.cuda()
            high_img = high_img.cuda()
            fake_img = model(low_img)
            for i in range(low_img.size(0)):
                gen_img = torch.clamp((fake_img[i].unsqueeze(0) + 1) / 2, 0, 1)
                gt_img = torch.clamp((high_img[i].unsqueeze(0) + 1) / 2, 0, 1)

                gen_np = gen_img.squeeze().cpu().numpy().transpose(1, 2, 0)
                gt_np = gt_img.squeeze().cpu().numpy().transpose(1, 2, 0)

                total_psnr += psnr(gt_np, gen_np, data_range=1.0)
                total_ssim += ssim(gt_np, gen_np, channel_axis=2, data_range=1.0)
                total_fsim += piq.fsim(gen_img, gt_img, data_range=1.0).item()
                total_mae += torch.abs(gen_img - gt_img).mean().item()
                count += 1
    return {
        'PSNR': total_psnr / count,
        'SSIM': total_ssim / count,
        'FSIM': total_fsim / count,
        'MAE': total_mae / count
    }

transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

train_dataset = LowLightDataset('/kaggle/input/lol-dataset/lol_dataset/our485/low', '/kaggle/input/lol-dataset/lol_dataset/our485/high', transform)
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)

gen = Generator().cuda()
disc = Discriminator().cuda()
vgg_loss = VGGPerceptual()
ssim_loss = piq.SSIMLoss().cuda()
lpips_model = lpips.LPIPS(net='alex').cuda()

opt_gen = torch.optim.Adam(gen.parameters(), lr=2e-4, betas=(0.5, 0.999))
opt_disc = torch.optim.Adam(disc.parameters(), lr=2e-4, betas=(0.5, 0.999))
bce = nn.BCELoss()
mae_loss = nn.L1Loss()

for epoch in range(100):
    gen.train()
    disc.train()
    for low_img, high_img in tqdm(train_loader):
        low_img = low_img.cuda()
        high_img = high_img.cuda()

        fake_img = gen(low_img)

        real_out = disc(high_img)
        fake_out = disc(fake_img.detach())

        real_label = torch.ones_like(real_out)
        fake_label = torch.zeros_like(fake_out)

        disc_loss = bce(real_out, real_label) + bce(fake_out, fake_label)
        opt_disc.zero_grad()
        disc_loss.backward()
        opt_disc.step()

        fake_out = disc(fake_img)
        adv_loss = bce(fake_out, real_label)
        rec_loss = mae_loss(fake_img, high_img)
        freq_loss = mae_loss(high_pass_filter(fake_img), high_pass_filter(high_img))
        perceptual_loss = lpips_model(fake_img, high_img).mean()

        features_fake = vgg_loss(fake_img)
        features_real = vgg_loss(high_img)
        vgg_feat_loss = nn.functional.l1_loss(features_fake, features_real)
        edge_val = edge_loss(fake_img, high_img)

        ssim_val = ssim_loss(torch.clamp((fake_img + 1) / 2, 0, 1),
                             torch.clamp((high_img + 1) / 2, 0, 1))

        total_gen_loss = (1.0 * adv_loss + 10.0 * rec_loss + 0.2 * freq_loss +
                          0.5 * perceptual_loss + 0.4 * vgg_feat_loss +
                          0.3 * edge_val + 0.5 * ssim_val)

        opt_gen.zero_grad()
        total_gen_loss.backward()
        opt_gen.step()

    metrics = compute_metrics(gen, train_loader)
    print(f"Epoch [{epoch+1}/100], Gen Loss: {total_gen_loss.item():.4f}, "
          f"Disc Loss: {disc_loss.item():.4f}, "
          f"PSNR: {metrics['PSNR']:.2f}, SSIM: {metrics['SSIM']:.4f}, "
          f"FSIM: {metrics['FSIM']:.4f}, MAE: {metrics['MAE']:.4f}")

# Visualize sample results
gen.eval()
samples = random.sample(range(len(train_dataset)), 4)
fig, axs = plt.subplots(4, 3, figsize=(12, 12))
for i, idx in enumerate(samples):
    low_img, high_img = train_dataset[idx]
    low_img_tensor = low_img.unsqueeze(0).cuda()
    with torch.no_grad():
        enhanced = gen(low_img_tensor).cpu().squeeze()
        enhanced = torch.clamp((enhanced + 1) / 2, 0, 1)

    axs[i, 0].imshow(torch.clamp((low_img + 1) / 2, 0, 1).permute(1, 2, 0))
    axs[i, 0].set_title("Low-light")
    axs[i, 1].imshow(enhanced.permute(1, 2, 0))
    axs[i, 1].set_title("Enhanced")
    axs[i, 2].imshow(torch.clamp((high_img + 1) / 2, 0, 1).permute(1, 2, 0))
    axs[i, 2].set_title("Ground Truth")

    for j in range(3):
        axs[i, j].axis('off')

plt.tight_layout()
plt.show()

torch.save(gen.state_dict(), 'generator.pth')
```

```python
import os
import glob
import torch
import numpy as np
from PIL import Image
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
import piq
import matplotlib.pyplot as plt
from tqdm import tqdm
import random

# --- Dataset ---
class LowLightDataset(Dataset):
    def __init__(self, low_dir, high_dir, transform=None):
        self.low_images = sorted(glob.glob(os.path.join(low_dir, '*')))
        self.high_images = sorted(glob.glob(os.path.join(high_dir, '*')))
        self.transform = transform

    def __len__(self):
        return len(self.low_images)

    def __getitem__(self, idx):
        low_img = Image.open(self.low_images[idx]).convert('RGB')
        high_img = Image.open(self.high_images[idx]).convert('RGB')
        if self.transform:
            low_img = self.transform(low_img)
            high_img = self.transform(high_img)
        return low_img, high_img

# --- Triple-Branch Generator ---
class Generator(torch.nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.structure_branch = torch.nn.Sequential(
            torch.nn.Conv2d(3, 64, 3, 1, 1), torch.nn.ReLU(),
            torch.nn.Conv2d(64, 64, 3, 1, 1), torch.nn.ReLU()
        )
        self.illum_branch = torch.nn.Sequential(
            torch.nn.Conv2d(3, 64, 3, 1, 1), torch.nn.ReLU(),
            torch.nn.Conv2d(64, 64, 3, 1, 1), torch.nn.ReLU()
        )
        self.detail_branch = torch.nn.Sequential(
            torch.nn.Conv2d(3, 64, 3, 1, 1), torch.nn.ReLU(),
            torch.nn.Conv2d(64, 64, 3, 1, 1), torch.nn.ReLU()
        )
        self.channel_attention = torch.nn.Sequential(
            torch.nn.AdaptiveAvgPool2d(1),
            torch.nn.Conv2d(192, 48, 1), torch.nn.ReLU(),
            torch.nn.Conv2d(48, 192, 1), torch.nn.Sigmoid()
        )
        self.output_layer = torch.nn.Sequential(
            torch.nn.Conv2d(192, 64, 3, 1, 1), torch.nn.ReLU(),
            torch.nn.Conv2d(64, 3, 3, 1, 1), torch.nn.Tanh()
        )

    def forward(self, x):
        s_feat = self.structure_branch(x)
        i_feat = self.illum_branch(x)
        d_feat = self.detail_branch(x)
        fused = torch.cat([s_feat, i_feat, d_feat], dim=1)
        att = self.channel_attention(fused)
        fused = fused * att
        return self.output_layer(fused)

# --- Evaluation ---
def compute_metrics(model, dataloader):
    model.eval()
    total_psnr = total_ssim = total_fsim = total_mae = 0
    count = 0
    with torch.no_grad():
        for low_img, high_img in tqdm(dataloader, desc="Evaluating"):
            low_img = low_img.cuda()
            high_img = high_img.cuda()
            fake_img = model(low_img)
            for i in range(low_img.size(0)):
                gen_img = torch.clamp((fake_img[i].unsqueeze(0) + 1) / 2, 0, 1)
                gt_img = torch.clamp((high_img[i].unsqueeze(0) + 1) / 2, 0, 1)

                gen_np = gen_img.squeeze().cpu().numpy().transpose(1, 2, 0)
                gt_np = gt_img.squeeze().cpu().numpy().transpose(1, 2, 0)

                total_psnr += psnr(gt_np, gen_np, data_range=1.0)
                total_ssim += ssim(gt_np, gen_np, channel_axis=2, data_range=1.0)
                total_fsim += piq.fsim(gen_img, gt_img, data_range=1.0).item()
                total_mae += torch.abs(gen_img - gt_img).mean().item()
                count += 1
    return {
        'PSNR': total_psnr / count,
        'SSIM': total_ssim / count,
        'FSIM': total_fsim / count,
        'MAE': total_mae / count
    }

# --- Config ---
low_dir = '/kaggle/input/lol-dataset/lol_dataset/eval15/low'
high_dir = '/kaggle/input/lol-dataset/lol_dataset/eval15/high'
model_path = 'generator.pth'
batch_size = 4

transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# --- Load Model & Data ---
test_dataset = LowLightDataset(low_dir, high_dir, transform)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

model = Generator().cuda()
model.load_state_dict(torch.load(model_path))
model.eval()

# --- Run Metrics ---
metrics = compute_metrics(model, test_loader)
print(f"\nTest Metrics:")
print(f"PSNR: {metrics['PSNR']:.2f}")
print(f"SSIM: {metrics['SSIM']:.4f}")
print(f"FSIM: {metrics['FSIM']:.4f}")
print(f"MAE:  {metrics['MAE']:.4f}")

# --- Visualization ---
fig, axs = plt.subplots(4, 3, figsize=(12, 12))
samples = random.sample(range(len(test_dataset)), 4)
for i, idx in enumerate(samples):
    low_img, high_img = test_dataset[idx]
    low_img_tensor = low_img.unsqueeze(0).cuda()
    with torch.no_grad():
        enhanced = model(low_img_tensor).cpu().squeeze()
        enhanced = torch.clamp((enhanced + 1) / 2, 0, 1)

    axs[i, 0].imshow(((low_img + 1) / 2).permute(1, 2, 0))
    axs[i, 0].set_title("Low-light")
    axs[i, 1].imshow(enhanced.permute(1, 2, 0))
    axs[i, 1].set_title("Enhanced")
    axs[i, 2].imshow(((high_img + 1) / 2).permute(1, 2, 0))
    axs[i, 2].set_title("Ground Truth")
    for j in range(3):
        axs[i, j].axis('off')
plt.tight_layout()
plt.show()
```
